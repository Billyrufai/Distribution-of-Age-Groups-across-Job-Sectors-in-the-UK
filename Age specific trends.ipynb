```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Creating a DateTimeIndex and loading neccesary files
```

```python
date = pd.date_range(start='2013-01-01', end='2023-12-31', freq='YE')
```

```python
df_16_24 = pd.read_excel('Data/16_24.xls',index_col = 'Title')
df_25_34 = pd.read_excel('Data/25_34.xls',index_col = 'Title')
df_35_49 = pd.read_excel('Data/35_49.xls',index_col = 'Title')
df_50_64 = pd.read_excel('Data/50_64.xls',index_col = 'Title')
df_65 = pd.read_excel('Data/65+.xls',index_col = 'Title')
```

```python
#Cleaning up column name
```

```python
df_16_24.columns = ['%']
df_16_24 = df_16_24.rename_axis('Year')
df_16_24.head(50)
```

```python
# I am only taking data for last 10 years so i will filter out the rest
```

```python
 df_16_24 = df_16_24.head(39)
```

```python
df_16_24
```

```python
df_16_24 = df_16_24.tail(11)
```

```python
df_16_24
```

```python
#Repeating for all age groups
```

```python
df_25_34.columns = ['%']
df_25_34 = df_25_34.rename_axis('Year')
df_25_34 = df_25_34.head(39)
df_25_34 = df_25_34.tail(11)
df_25_34
```

```python
df_35_49.columns = ['%']
df_35_49 = df_35_49.rename_axis('Year')
df_35_49 = df_35_49.head(39)
df_35_49 = df_35_49.tail(11)
df_35_49
```

```python
df_50_64.columns = ['%']
df_50_64 = df_50_64.rename_axis('Year')
df_50_64 = df_50_64.head(39)
df_50_64 = df_50_64.tail(11)
df_50_64
```

```python
df_65.columns = ['%']
df_65 = df_65.rename_axis('Year')
df_65 = df_65.head(39)
df_65 = df_65.tail(11)
df_65
```

```python
df_16_24.index = ['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
df_25_34.index = ['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
df_35_49.index = ['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
df_50_64.index = ['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
df_65.index = ['2013','2014','2015','2016','2017','2018','2019','2020','2021','2022','2023']
```

```python
#Combining dataframes now
```

```python
df = pd.concat([df_16_24.rename(columns={'%': '16-24'}),df_25_34.rename(columns={'%': '25-34'}),df_35_49.rename(columns={'%': '35-49'}),df_50_64.rename(columns={'%': '50-64'}),df_65.rename(columns={'%': '65+'})], axis=1) 
```

```python
df
```

```python
#plotting it using year as x axis and y as the values for each age group
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df.index,df['16-24'], label='Age 16-24')
plt.plot(df.index,df['25-34'], label='Age 25-34')
plt.plot(df.index,df['35-49'], label='Age 35-49')
plt.plot(df.index,df['50-64'], label='Age 50-64')
plt.plot(df.index,df['65+'], label='Age 65+')
plt.title('Time Series Data for Different Age Groups')
plt.xlabel('Year')
plt.ylabel('Employment%')
plt.tight_layout()
plt.legend(title="Age Group", bbox_to_anchor=(1.05, 1), loc='upper left')
```
