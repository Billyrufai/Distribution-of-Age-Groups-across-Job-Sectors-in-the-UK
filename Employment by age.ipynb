```python
#Creating a stacked bar graph to visualise the dataframe
```

```python
df1_bar = df1.plot(kind='bar', stacked=True, figsize=(14, 8), colormap='tab20')
plt.title("Employment by Age Group and Industry")
plt.xlabel("Industry")
plt.ylabel("Number of Employees")
plt.legend(title="Age Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
df2_bar= df2.plot(kind='bar', stacked=True, figsize=(14, 12), colormap='tab20')
plt.title("Employment by Age Group and Industry")
plt.xlabel("Industry")
plt.ylabel("Number of Employees")
plt.legend(title="Age Group", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
