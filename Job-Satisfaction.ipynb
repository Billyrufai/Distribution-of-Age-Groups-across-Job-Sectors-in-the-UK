```python

import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('Data/Job-satisfaction.xlsx')
```

```python
df
```

```python
#Removing missing values
```

```python
df = df.dropna()
```

```python
df
```

```python
#Removing uneccesary columns
```

```python
df = df.drop(columns = ['Unnamed: 1','Unnamed: 2'])
```

```python
df
```

```python
#Renaming column names and reseting index
```

```python
df.columns = ['Age Group','%']
```

```python
df
```

```python
df.drop(5,inplace = True)
```

```python
df.reset_index(inplace = True)
```

```python
df
```

```python
df =df.drop(columns='index')
```

```python
df
```

```python
#Creating  bar chart
```

```python
plt.figure(figsize=(10, 6))
plt.bar(df['Age Group'], df['%'], color='skyblue')
plt.xlabel('Age Group')
plt.ylabel('Population%')
plt.title('Population% of age groups working satisfactory hours')
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```
